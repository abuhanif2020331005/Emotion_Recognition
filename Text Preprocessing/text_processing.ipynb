Handling emoji

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # Transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
text = "I love Python üòçüî•üíØ"
text=remove_emoji(text)
text
```

Lemmatization

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag, word_tokenize

lemmatizer = WordNetLemmatizer()

# Helper: Convert POS tag from Treebank to WordNet
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun

# Example
text = "The children are playing in the gardens"
tokens = word_tokenize(text)
tagged = pos_tag(tokens)

lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in tagged]

print(" ".join(lemmatized_words))
# Output: The child be play in the garden
```

```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "The children were playing with their toys in the park."

# Process the text
doc = nlp(text)

# Extract lemmas (excluding punctuation)
lemmatized_words = [token.lemma_ for token in doc if not token.is_punct]

# Print result
print("Original :", text)
print("Lemmatized:", " ".join(lemmatized_words))
```
